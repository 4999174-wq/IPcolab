```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/lena256.jpg')
print(img.shape)
gImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gImg, cmap='gray')
```

```python
def saturation(value):
  if (value > 255):
    value = 255
  elif (value < 0):
    value = 0
  return value
```

```python
kernal3 = np.ones((3,3), np.float32)/9
kernal5 = np.ones((5,5), np.float32)/25
kernal7 = np.ones((7,7), np.float32)/49
```

```python
output = cv2.filter2D(gImg, -1, kernal3)
cv2_imshow(output)
```

```python
output = cv2.filter2D(gImg, -1, kernal5)
cv2_imshow(output)
```

```python
output = cv2.filter2D(gImg, -1, kernal7)
cv2_imshow(output)
```

```python
gOut = np.zeros((gImg.shape[0], gImg.shape[1]), dtype=np.ubyte)
```

```python
for h in range(1, gImg.shape[0]-1):
  for w in range(1, gImg.shape[1]-1):
    sum = 0
    for i in range(-1,2):
      for j in range(-1,2):
        sum += gImg[h+i][w+j]*kernal3[i+1][j+1]
    sum = saturation(sum)
    gOut[h][w] = sum

cv2_imshow(gOut)
```

```python
output = cv2.filter2D(output, -1, kernal3)
cv2_imshow(output)
```
