```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/girl.jpg')
print(img.shape)
gImg1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gImg1, cmap='gray')
```

```python
hist = cv2.calcHist([gImg1], [0], None, [256], [0, 256])
plt.plot(hist)
```

```python
outImg = np.zeros(gImg1.shape, dtype='uint8')
```

```python
min = 0
max = 200

for i in range(gImg1.shape[0]):
  for j in range(gImg1.shape[1]):
    if gImg1[i][j] < min:
      outImg[i][j] = 0
    elif gImg1[i][j] > max:
      outImg[i][j] = 255
    else:
      val = np.int32(gImg1[i][j] - min)
      val = val * 255 / (max - min)
      outImg[i][j] = val

cv2_imshow(outImg)
```

```python
hist2 = cv2.calcHist([outImg], [0], None, [256], [0, 256])
plt.plot(hist2)
```
