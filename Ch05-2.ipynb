```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/gsy2.jpg')
print(img.shape)
gImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gImg)
```

```python
outImg = cv2.blur(gImg, (7,7))
cv2_imshow(outImg)
```

```python
outImg = cv2.GaussianBlur(gImg, (7,7), 5)
cv2_imshow(outImg)
```

```python
outImg = cv2.medianBlur(gImg, 7)
cv2_imshow(outImg)
```

```python
outImage = gImg.copy()
#blurImage = cv2.GaussianBlur(outImage, (7,7), 5)
blurImage = cv2.medianBlur(outImage, 7)

x = 200 + 70
y = 200 + 65
size = 15

for h in range(y, y+size):
  for w in range(x, x+size):
    #outImage[h][w] = 255
    outImage[h][w] = blurImage[h][w]

cv2_imshow(outImage)
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/lena256.jpg')
print(img.shape)
gImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#plt.imshow(gImg, cmap='gray')
cv2_imshow(gImg)
```

```python
outImg = cv2.GaussianBlur(gImg, (7,7), 2)
cv2_imshow(outImg)
```

```python
edgeImage = cv2.addWeighted(gImg, 1, outImg, -1, 0)
cv2_imshow(edgeImage)
```

```python
unsharpImage = cv2.addWeighted(gImg, 1, edgeImage, 2, 0)
cv2_imshow(unsharpImage)
```

```python
kernal3 = np.ones((3,3), np.float32)/9
kernal3[0][0] = -1/9
kernal3[0][2] = -1/9
kernal3[2][0] = -1/9
kernal3[2][2] = -1/9
kernal3[1][0] = -1/9
kernal3[1][2] = -1/9
kernal3[0][1] = -1/9
kernal3[2][1] = -1/9
kernal3[1][1] = 17/9
```

```python
output = cv2.filter2D(gImg, -1, kernal3)
cv2_imshow(output)
```
